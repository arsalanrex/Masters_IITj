```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data with proper date parsing and specifying date format
data = pd.read_csv("5y_data.csv", parse_dates=['date'], dayfirst=True, date_format='%d/%m/%y')

# Drop the 'date' column
data = data.drop(columns=['date'])

# Display the first few rows of the data
print(data.head())

# Basic statistical measures
print(data.describe())

# Calculate correlation coefficients
correlation = data.corr()
print(correlation['GOLD-1-MCX'])

# Plot heatmap of correlation coefficients
plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Create scatter plots
plt.figure(figsize=(12, 8))
plt.subplot(2, 3, 1)
plt.scatter(data['GOLD-1-MCX'], data['COPPER-1-MCX'])
plt.title('Gold vs Copper')
plt.xlabel('Gold Price')
plt.ylabel('Copper Price')

plt.subplot(2, 3, 2)
plt.scatter(data['GOLD-1-MCX'], data['CRUDEOIL-1-MCX'])
plt.title('Gold vs Crude Oil')
plt.xlabel('Gold Price')
plt.ylabel('Crude Oil Price')

plt.subplot(2, 3, 3)
plt.scatter(data['GOLD-1-MCX'], data['SILVER-1-MCX'])
plt.title('Gold vs Silver')
plt.xlabel('Gold Price')
plt.ylabel('Silver Price')

plt.subplot(2, 3, 4)
plt.scatter(data['GOLD-1-MCX'], data['NATURALGAS-1-MCX'])
plt.title('Gold vs Natural Gas')
plt.xlabel('Gold Price')
plt.ylabel('Natural Gas Price')

plt.subplot(2, 3, 5)
plt.scatter(data['GOLD-1-MCX'], data['GUARSEED-1-MCX'])
plt.title('Gold vs Guar Seed')
plt.xlabel('Gold Price')
plt.ylabel('Guar Seed Price')

plt.tight_layout()
plt.show()
```
