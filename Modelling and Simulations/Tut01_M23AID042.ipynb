**Implement Gauss-Jacobi method and Gauss-Seidel method in Python. You need it to solve the
next problem.**

```python
import numpy as np 

def gauss_jacobi(A, b, x0, tol = 1e-6, max_iter=1000):
    n = len(b)
    x = x0.copy()
    x_prev = np.zeros_like(x)
    
    for k in range(max_iter) : 
        for i in range(n):
            x_prev[i] = x[i]
            sigma = 0
            for j in range(n):
                if j != i:
                    sigma += A[i, j] * x_prev[j]
            x[i] = (b[i] - sigma) / A[i, i]
        
        if np.linalg.norm(x - x_prev, 2) < tol:
            return x, k + 1
    return x, max_iter

def gauss_seidel(A, b, x0, tol = 1e-6, max_iter=1000):
    n = len(b)
    x = x0.copy()
    for k in range(max_iter) : 
        for i in range(n):
            sigma = 0
            for j in range(n):
                if j != i:
                    sigma += A[i, j] * x[j]
            x[i] = (b[i] - sigma) / A[i, i]
        if np.linalg.norm(np.dot(A, x) - b, 2) < tol:
            return x, k + 1
    return x, max_iter

A = np.array([[10, -1, 2], 
              [1, 10, -1], 
              [-2, 1, 10]], dtype=float)
b = np.array([6, 9, 6],  dtype=float)
x0 = np.zeros_like(b)

result_jacobi, iterations_jacobi = gauss_jacobi(A, b, x0)
print("\nGauss-Jacobi:")
print ("Solution:", [round(num, 4) for num in result_jacobi])
print ("Iterations:", iterations_jacobi)

result_seidel, iterations_seidel = gauss_seidel(A, b, x0)
print("\nGauss-Seidel:")
print ("Solution:", [round(num, 4) for num in result_seidel])
print ("Iterations:", iterations_seidel)
```

**Modeling in action: An oil company wants build 25 cuboid tanks with a combined capacity of 28,000 cubic meter. Three different types of tank designs are under consideration: a 10-meter wide tank with a capacity of 350 cubic meter, a 14-meter wide tank with a capacity of 700 cubic meter, and a 24-meter tank with a capacity of 1,400 cubic meter. How many of each type of tank the company can build?**

```python
import numpy as np

# Define tank designs and their capacities
tank_designs = np.array([
    [10, 350],
    [14, 700],
    [24, 1400]
])

total_tanks = 25
total_capacity = 28000
min_tanks_each = 0  # use 1 if at least one tank for each size is required

# Generate all possible combinations of tanks
combinations = np.array(np.meshgrid(*[range(min_tanks_each, total_tanks + 1)] * len(tank_designs))).T.reshape(-1, len(tank_designs))

# Filter combinations to meet capacity requirement and total tanks limit
valid_combinations = combinations[((combinations @ tank_designs[:, 1]) == total_capacity) & (np.sum(combinations, axis=1) == total_tanks)]

# Check for valid combinations
if len(valid_combinations) > 0:
    print("Solutions:")
    for combination in valid_combinations:
        print(f"{combination[0]} tanks of 10m, {combination[1]} tanks of 14m, and {combination[2]} tanks of 24m.")
else:
    print("No solution found.")
    
# since there are no hidden variable we cannot optimize the solution. 
# There is no minimum number of tank for each type listed in the problem statement so I have taken that as 0.
# The solution is not unique, there are multiple ways to achieve the goal.
```

**Pick a city of your choice. Pick a year of your choice. Get the daily maximum and minimum temperature data for that year for that city. So you should have about 365 data points for each series. Analyse the temperature trend using the curve fitting technique. Mention your data source.**

```python
# We are using meteostat library to get the temperature data for Gaya, Bihar, India for the year 2023. Meteostat is a Python library designed for retrieving historical weather and climate data from the Meteostat project's database

from datetime import datetime
from meteostat import Point, Daily
import numpy as np
from scipy.optimize import curve_fit
import plotly.graph_objects as go

# Set time period
start = datetime(2023, 1, 1)
end = datetime(2023, 12, 31)

# Create Point for Gaya, Bihar, India (my hometown)
gaya = Point(24.7914, 85.0002)

# Get daily data for 2023
data = Daily(gaya, start, end)
data = data.fetch()

selected_data = data[['tmin', 'tmax']]

# Define Gaussian function
def gaussian(x, A, mu, sigma):
    return A * np.exp(-(x - mu)**2 / (2 * sigma**2))

# Fit Gaussian curve to tmin
popt_tmin, pcov_tmin = curve_fit(gaussian, np.arange(len(selected_data)), selected_data['tmin'], p0=[1, len(selected_data)/2, 1])

# Fit Gaussian curve to tmax
popt_tmax, pcov_tmax = curve_fit(gaussian, np.arange(len(selected_data)), selected_data['tmax'], p0=[1, len(selected_data)/2, 1])

# Generate x values for plotting
x = np.linspace(0, len(selected_data), 100)

# Plotting with Plotly
fig = go.Figure()

# Scatter plot for tmin
fig.add_trace(go.Scatter(x=np.arange(len(selected_data)), y=selected_data['tmin'], mode='markers', name='tmin', marker=dict(color='blue')))

# Scatter plot for tmax
fig.add_trace(go.Scatter(x=np.arange(len(selected_data)), y=selected_data['tmax'], mode='markers', name='tmax', marker=dict(color='red')))

# Gaussian fit for tmin
fig.add_trace(go.Scatter(x=x, y=gaussian(x, *popt_tmin), mode='lines', name='tmin fit', line=dict(color='blue', dash='dash')))
# Gaussian fit for tmax
fig.add_trace(go.Scatter(x=x, y=gaussian(x, *popt_tmax), mode='lines', name='tmax fit', line=dict(color='red', dash='dash')))

fig.update_layout(
    title='Curve Fitting for Min and Max Temperature for Gaya, Bihar, India (2023)',
    xaxis_title='Day',
    yaxis_title='Temperature',
    legend=dict(x=0, y=1.0, bgcolor='rgba(255, 255, 255, 0.5)'),
    plot_bgcolor='white',
    xaxis=dict(showgrid=True, gridcolor='lightgrey'),
    yaxis=dict(showgrid=True, gridcolor='lightgrey')
)

fig.show()

```

```python
import numpy as np

# Data: Student scores
data = {
    'Sr. No.': [428, 215, 221, 159, 247, 146, 222, 449, 225, 257, 299, 315, 329, 460, 507, 171, 232, 249, 389, 484, 485, 506, 63, 141, 280, 324, 492, 400, 34],
    'T1': [17, 15, 16, 15, 14, 15, 12, 14, 18, 17, 13, 10, 13, 15, 9, 14, 13, 13, 14, 9, 12, 15, 13, 15, 2, 0, 11, 8, 3],
    'T2': [15, 18, 20, 16, 15, 17, 11, 13, 18, 13, 15, 12, 15, 10, 16, 14, 15,17, 10, 9, 8, 7, 8, 12, 10, 0, 3, 13, 1],
    'T3': [28, 30,	23,	28,	29,	25,	29,	28,	25,	23,	20,	24,	20,	26,	20,	25,	15,	23,	18,	24,	19,	16,	15,	17,	17,	22,	16, 8, 18],
    'T4': [29, 27, 31, 28, 32, 31, 24, 23, 23, 26, 29, 27,	28,	22,	27,	22,	28,	21,	19,	16,	20,	20,	23,	22,	14,	9, 12, 13, 12],
    'G': [10, 10, 10, 10, 10, 10, 9, 9,	9, 9, 9, 9,	9, 8, 8, 8, 8, 8, 7, 7, 7, 7, 7, 7,	6, 6, 6, 6,	6]
}

# Convert data to numpy array
data_array = np.array([data['T1'], data['T2'], data['T3'], data['T4']]).T

# Grade points
grades = np.array(data['G'])

# Perform SVD
U, S, Vt = np.linalg.svd(data_array, full_matrices=False)

# Define a function to recommend grade points for new students
def recommend_grade(new_student_scores):
    # Project new student scores onto the principal components
    projected_scores = np.dot(new_student_scores, Vt.T)
    # Find the nearest neighbor in the principal component space
    nearest_neighbor_index = np.argmin(np.linalg.norm(projected_scores - np.dot(data_array, Vt.T), axis=1))
    # Return the grade of the nearest neighbor
    return grades[nearest_neighbor_index]

# Test the recommendation function
students_to_recommend = [
    [127, 10, 7, 15, 20],
    [365, 13, 15, 25, 18],
    [210, 18, 18, 28, 27],
    [34, 3, 1, 18, 12]
]

for student_scores in students_to_recommend:
    student_id = student_scores[0]
    recommended_grade = recommend_grade(student_scores[1:])
    print(f"For Sr. No. {student_id}, recommended grade: {recommended_grade}")
```
